```python
#The data contains 12 columns and 891 rows. 
#To start the cleansing of data we start with counting the null values in the data
df.isnull().sum()
```

```python
# As we can see there are around 2 null values in column 'Embarked' and 177 in 'Age' and 687 in 'Cabin'.
# The columns such as Sibsp,Parch,Ticket,Cabin,Embarked doesnot hold any logical importance within the dataset,
# the survival doenot not depend on the values of these columns hence it would be a good approach to remove the
# unnecessary columns.
```

```python
df2 = df.drop(['Cabin','Ticket','SibSp','Id','Name','Embarked','Parch'],axis = 'columns')
df2.head()
```

```python
# Now to fill in the nas value in age we did some digging into the data to understand the pattern of age w.r.t Pclass No.
# We performed some basic data exploratory operations and came to the conclusion that the maximum nas in the age are from
# PclassNo. 3 followed by 1 and then 2.
# So we explored some age patterns within each PclassNo and came to the conclusion that maximum age in the PclassNo. 3 is 28 and
# Pclassno. 2 is and Pclassno.1 is
# So we filled in nas in age accordingly.
```

```python
def fill_nas_in_age(df):
    df_without_nas = pd.DataFrame()
    df_with_nas = pd.DataFrame()
    check  = pd.DataFrame()
    
    df_with_nas_class3 = df[df['PclassNo']==3][df['Age'].isnull()]
    df_with_nas_class2 = df[df['PclassNo']==2][df['Age'].isnull()]
    df_with_nas_class1 = df[df['PclassNo']==1][df['Age'].isnull()]
    df_without_nas = df[~df['Age'].isnull()]
    df_with_nas_class3['Age'] = 25 #mean age in class 3
    df_with_nas_class2['Age'] = 30 #mean age in class 2
    df_with_nas_class1['Age'] = 38 #mean age in class 1
    
    df_without_nas = pd.concat([df_with_nas_class3,df_without_nas])
    df_without_nas = pd.concat([df_with_nas_class2,df_without_nas])
    df_without_nas = pd.concat([df_with_nas_class1,df_without_nas])
    
    return df_without_nas
```

```python
df3 = fill_nas_in_age(df2)
df3
```

```python
df3.isnull().sum()
```

```python
#As all the nas are removed its time to change all the non-categorical data into categorical for the best fit of model.
```

```python
def simplify_ages(df):
    bins = (-1,0,5,12,18,25,35,60)
    group_names = ['Unknown','Baby','Child','Teenager','Young Adult','Adult','Senior']
    categories = pd.cut(df.Age,bins,labels=group_names)
    df.Age = categories.astype(str)
    return df

def simplify_fare(df):
    bins = (0,8,14,31,512)
    group_names=['1_quartile','2_quartile','3_quartile','High Fare']
    categories = pd.cut(df.Fare,bins,labels=group_names)
    df.Fare = categories.astype(str)
    return df
```

```python
df4 = simplify_ages(df3)
df5 = simplify_fare(df4)
df5.head()
```

```python
#The creation of dummy variables will ease and speed up the performance of the algorithms used for classification.
```

```python
def Concat_df(df):
    dummies_Sex = pd.get_dummies(df.Sex)
    dummies_Age = pd.get_dummies(df.Age)
    dummies_Fare =pd.get_dummies(df.Fare)
    df = pd.concat([df,dummies_Sex],axis='columns')
    df = pd.concat([df,dummies_Age],axis = 'columns')
    df = pd.concat([df,dummies_Fare],axis = 'columns')
    return df
```

```python
df6 = Concat_df(df5)
```

```python
df7 = df6.drop(['Sex','Age','Fare','nan'],axis = 'columns')
```

```python
x = df7.drop('Survived',axis = 'columns')
```

```python
y = df7.Survived
```

```python
# We have divided the data into the train and test and applied a gridsearch method to identify the best classification algorithm
# that works with the dataset
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.model_selection import ShuffleSplit
def find_best_model_using_gridsearch(x,y):
    algos = {
        'logistic_regression':{
            'model':LogisticRegression(),
            'params':{
                'C':np.logspace(-3,3,7),
                'penalty':["l2"],
                'max_iter':[100],
                'fit_intercept':[True,False]
            }
        },
        'naive_bayes_Multi':{
            'model':MultinomialNB(),
            'params':{
                'fit_prior':[True,False],
                'alpha':(1,0.1,0.01)
                
            }
        },
        'Random_forest':{
            'model':RandomForestClassifier(),
            'params':{
            'n_estimators':[4,6,9],
            'max_features':['log2','sqrt','auto'],
            'criterion':['entropy','gini'],
            'max_depth':[2,3,5,10,12,15],
            'min_samples_split':[2,3,5],
            'min_samples_leaf':[1,5,8]
            }
        },
        'Decision_Tree':{
            'model':tree.DecisionTreeClassifier(),
            'params':{
                'criterion':['entropy','gini'],
                'splitter':['best','random'],
                'max_depth':[2,3,5,8,12,16],
                'min_samples_split':[2,3,5],
                'min_samples_leaf':[1,5,8]
            }
        },
        'naive_bayes_Gaussian':{
            'model':GaussianNB(),
            'params':{
                
                'var_smoothing':[1e-09]
                
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits = 5,test_size=0.2,random_state=0)
    for algo_name,config in algos.items():
        gs = GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params': gs.best_params_
    })
    
    return pd.DataFrame(scores,columns=['model','best_score',"best_params"])
    
```

```python
find_best_model_using_gridsearch(x,y)
```
